```python
import json
from pathlib import Path

file_path = Path("data_json/documents.jsonl")

docs = []
with open(file_path, 'r', encoding='utf-8') as f:
    for i, line in enumerate(f):
        doc = json.loads(line)
        
        if i < 5:
            print(f"Sample Document {i}:", doc)
        
        if i % 1000 == 0:
            print(f"Processed {i} lines...", end='\r')
```
