```python
import ir_datasets
import json
import os
from itertools import islice
```

```python
dataset = ir_datasets.load("cord19/trec-covid/round1")
```

```python
# documents metadata
print(dataset.docs_cls()._fields)
```

```python
for doc in islice(dataset.docs_iter(), 3): # look at first 3 documents
    print(f"ID: {doc.doc_id}")    
    print(f"Title: {doc.title}")
    print(f"Abstract: {doc.abstract}")
    print(f"Link: {doc.doi}")
    print(f"date: {doc.date}")
    print("-" * 20)
```

```python
doc_store = dataset.docs_store()
query_store = {q.query_id: q for q in dataset.queries_iter()}

count = 0
for qrel in dataset.qrels_iter():
    if qrel.relevance == 2:
        print("query relevance == 2")
        query = query_store[qrel.query_id]
        doc = doc_store.get(qrel.doc_id)
        
        print(f"SAMPLE {count + 1}")
        print(f"QUERY ID:    {query.query_id}")
        print(f"QUERY TITLE: {query.title}")
        print(f"QUERY DESC:  {query.description}")
        print("-" * 30)
        print(f"MATCHING DOC ID: {doc.doc_id}")
        print(f"DOC TITLE:       {doc.title}")
        print(f"DOC ABSTRACT:    {doc.abstract[:300]}...") # Truncated for readability
        print("\n" + "="*50 + "\n")
        
        count += 1
        if count >= 6:
            break
    
    elif qrel.relevance == 1:
        print("query relevance == 1")
        query = query_store[qrel.query_id]
        doc = doc_store.get(qrel.doc_id)
        
        print(f"SAMPLE {count + 1}")
        print(f"QUERY ID:    {query.query_id}")
        print(f"QUERY TITLE: {query.title}")
        print(f"QUERY DESC:  {query.description}")
        print("-" * 30)
        print(f"MATCHING DOC ID: {doc.doc_id}")
        print(f"DOC TITLE:       {doc.title}")
        print(f"DOC ABSTRACT:    {doc.abstract[:300]}...") # Truncated for readability
        print("\n" + "="*50 + "\n")

        count += 1
        if count >=3:
            continue
            
    
```

```python
from collections import Counter
import ir_datasets

qrel_counts = Counter(q.query_id for q in dataset.qrels_iter())

print(f"{'Query ID':<10} | {'Total Judgments':<15}")
print("-" * 30)

for q_id in sorted(qrel_counts.keys(), key=int):
    print(f"{q_id:<10} | {qrel_counts[q_id]:<15}")
```
