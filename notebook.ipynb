```python
from pyserini.search.lucene import LuceneSearcher
import json
from retrieval.BM25 import search_trec_covid
from retrieval.BM25F import search_trec_covid_bm25f
import ir_datasets
```

```python
dataset = ir_datasets.load("cord19/trec-covid/round1")
```

```python
# documents metadata
print(dataset.docs_cls()._fields)
```

```python
searcher = LuceneSearcher('indexes/trec_covid_lemmatized')
searcher.set_bm25(k1=1.2, b=0.75)
```

```python
results = search_trec_covid("I have covid, will I also get cancer?")

for i, hit in enumerate(results):
        doc = searcher.doc(hit.docid)
        doc_data = json.loads(doc.raw())
        title = doc_data.get('title', 'No Title Found')
        
        print(f"Rank:  {i+1}")
        print(f"ID:    {hit.docid}")
        print(f"Score: {hit.score:.4f}")
        print(f"Title: {title}")
        print("-" * 30)
```

```python
results = search_trec_covid_bm25f("covid-19 children pediatric")

for i, hit in enumerate(results):
        doc = searcher.doc(hit.docid)
        doc_data = json.loads(doc.raw())
        title = doc_data.get('title', 'No Title Found')
        
        print(f"Rank:  {i+1}")
        print(f"ID:    {hit.docid}")
        print(f"Score: {hit.score:.4f}")
        print(f"Title: {title}")
        print("-" * 30)
```

```python
import spacy
nlp = spacy.load("en_core_web_sm", disable=["parser", "ner"])
```

```python
def view_query(query):
    tokens = [t.lemma_.lower() for t in nlp(query) if not t.is_stop and not t.is_punct]
    clean_query = " ".join(tokens)

    return f"Query: '{query}' -> Cleaned query: '{clean_query}'"
```

```python
view_query("covid.... kids")
```

```python
view_query("children organising covid-19")
```
