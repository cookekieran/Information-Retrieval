```python
nlp = spacy.load("en_core_web_sm", disable=["parser", "ner", "attribute_ruler"])
```

```python
def preprocess_and_save(filename="cleaned_corpus.pkl"):
    processed_data = {}
    
    # We pull doc_id, title, and abstract
    # We use a generator to keep memory usage low
    def doc_generator():
        for doc in dataset.docs_iter():
            # Combine title and abstract for processing, or keep separate for BM25F
            # Here we keep them separate to allow for BM25F weighting later
            yield (doc.doc_id, doc.title or "", doc.abstract or "")

    print("Starting spaCy pipeline (this may take a few minutes)...")
    
    # 3. Use nlp.pipe for high-speed batching
    # We process titles and abstracts separately to maintain "Field" logic
    count = 0
    for doc_id, title, abstract in tqdm(doc_generator(), total=49790):
        # Process title
        t_doc = nlp(title.lower())
        t_tokens = [token.lemma_ for token in t_doc if not token.is_stop and not token.is_punct and token.is_alpha]
        
        # Process abstract
        a_doc = nlp(abstract.lower())
        a_tokens = [token.lemma_ for token in a_doc if not token.is_stop and not token.is_punct and token.is_alpha]
        
        processed_data[doc_id] = {
            "t": t_tokens,
            "a": a_tokens
        }
        
    # 4. Pickle the results to a file
    print(f"Saving cleaned data to {filename}...")
    with open(filename, "wb") as f:
        pickle.dump(processed_data, f)
    print("Done!")

if __name__ == "__main__":
    preprocess_and_save()
```
